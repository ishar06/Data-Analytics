```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
heights_no_outlier = [160, 162, 163, 165, 166, 167, 168, 168, 169, 170, 171, 172, 173, 174, 175]

heights_with_outlier = [140, 160, 162, 163, 165, 166, 167, 168, 168, 169, 170, 171, 172, 190, 195]

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].boxplot(heights_no_outlier, orientation='vertical')
axes[0].set_title("Class A (No Outliers)")
axes[0].set_ylabel("Height (cm)")

axes[1].boxplot(heights_with_outlier, orientation='vertical')
axes[1].set_title("Class B (With Outliers)")
axes[1].set_ylabel("Height (cm)")

plt.suptitle("Boxplot of Student Heights (2 Classes)", fontsize=14, color="purple")
plt.show()
```

```python
students = ['A', 'B', 'C', 'D', 'E']
math_scores = [88, 92, 79, 93, 85]
science_scores = [75, 89, 95, 80, 70]


fig,ax = plt.subplots(figsize=(10,5))
x = np.arange(len(students)) 
width = 0.35
ax.bar(x - width/2, math_scores, width, label='Math', color="black")
ax.bar(x + width/2, science_scores, width, label='Science', color="red")
ax.set_xlabel("Students")
ax.set_ylabel("Scores")
ax.set_title("Math vs Science Scores")
ax.set_xticks(x)
ax.set_xticklabels(students)
ax.legend()
plt.show()
```
