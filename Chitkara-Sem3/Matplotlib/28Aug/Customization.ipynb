```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

years = np.arange(2015, 2025)
admissions = np.array([45, 59, 65, 83, 87, 70, 81, 71, 90, 72])

min_year, max_year = years.min(), years.max()
min_adm, max_adm = admissions.min(), admissions.max()

fig, ax = plt.subplots(figsize=(10, 5))


ax.plot(years, admissions, marker='o', label='Admissions')

ax.set_xlim(min_year - 1, max_year + 1)
ax.set_ylim(min_adm - 10, max_adm + 10)

year_max, val_max = years[admissions.argmax()], admissions[admissions.argmax()]
year_min, val_min = years[admissions.argmin()], admissions[admissions.argmin()]


plt.annotate(
    f"Highest\n {year_max} : {val_max}",
    xy = (year_max, val_max),
    xytext = (year_max, val_max + 11),
    arrowprops = dict(facecolor="green", edgecolor="green", arrowstyle="->", linewidth=1.5),
    color = "green",
    fontsize = 10,
    ha="center"
)

plt.annotate(
    f"Lowest\n {year_min} : {val_min}",
    xy = (year_min, val_min),
    xytext = (year_min, val_min + 10),
    arrowprops = dict(facecolor="black", edgecolor="red", arrowstyle="->", linewidth=1.5),
    color = "red",
    fontsize = 10,
    ha="center"
)

for x,y in zip(years, admissions):
    plt.text(x - 0.5, y - 3, f"({x}:{y})", fontsize=8, color="maroon")


ax.set_xlabel('Year')
ax.set_ylabel('Number of Admissions')
ax.set_title('Admissions per Year (20152024)')

ax.grid(True)
ax.legend()
plt.show()
```

```python
from matplotlib import colormaps

countries = ["Argentina","Sweden", "South Africa", "Spain", "China", "Uganda", "Japan", "India", "Maurtius"]
densities = [44, 67, 140, 250, 390, 640, 880, 1260, 1650]

plt.subplots(figsize=(10,6))
sc = plt.scatter(countries, densities, c=densities, cmap="viridis", edgecolor="black", s=200)

cbar = plt.colorbar(sc)
cbar.set_label("Population Density (per km2)", fontsize=12)

for i in range(len(densities)):
    plt.text(countries[i], densities[i]+50, densities[i], fontsize=8, ha="center", color="black")


plt.grid(True)
plt.show()
```

```python
import numpy as np
from matplotlib import colormaps
viridis = colormaps['viridis']

values = np.linspace(0,1,6)
colors = [viridis(v) for v in values]

plt.figure(figsize=(8,1))
for i, c in enumerate(colors):
    plt.fill_between([i,i+1],0,1, color=c)
    plt.text(i+0.5, 0.5, f"{values[i]:.2f}", ha="center", va="center", color="white", fontsize=12)

plt.axis("off")
plt.show()
```
