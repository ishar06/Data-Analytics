```python
fig, axes = plt.subplots(1, 2, figsize=(8,3))
# 1 --> no. of rows, 2 --> no. of columns

#Left Plot
axes[0].plot([1,2,3,4,5,6,7,8,9],[4,6,9,16,25,36,49,64,81])
axes[0].set_title("Quadratic")

#Right Plot
axes[1].plot([1,2,3],[1,2,3])
axes[1].set_title("Linear")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

marketing_urban = np.array([10, 15, 20, 25, 30])
sales_urban = np.array([100, 150, 200, 230, 280])

marketing_rural = np.array([5, 10, 15, 20, 25])
sales_rural = np.array([60, 90, 130, 160, 190])

# Plotting
fig, ax = plt.subplots(figsize=(7,4))
ax.scatter(marketing_urban, sales_urban, c="red", marker="o", label="Urban")
ax.scatter(marketing_rural, sales_rural, c="orange", marker="^", label="Rural")

ax.set_xlabel("Marketing Expense ($k)")
ax.set_ylabel("Sales Expense ($)")
ax.set_title("Sales vs Marketing Expenses")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ["Urban", "Rural"]

sales = [250, 180]

fig, ax = plt.subplots(figsize=(6,4))
ax.bar(categories, sales, color=["red", "orange"])

ax.set_xlabel("Region")
ax.set_ylabel("Sales ($k)")
ax.set_title("Urban vs Rural Sales")
plt.show()
```

```python
data = np.random.normal(0,1,1000)
# 0 --> mean
# 1 --> standard deviation
# 1000 --> no. of values

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
sizes = [20,30,25,25]
labels = ['Python','Java','C++','JS']

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Pie Chart")
plt.show()
```
