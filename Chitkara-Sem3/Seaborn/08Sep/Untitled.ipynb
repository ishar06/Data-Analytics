```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(-5, 5, 100)
y = x**2

data = np.random.randn(1000)

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
colors = np.random.rand(50)

fig, axes = plt.subplots(2,2, figsize = (12,6))
axes[0,0].plot(x, y, color="blue")
axes[0, 0].set_title("Line Plot: y = x")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("y")
axes[0, 0].set_xlim(-5, 5)
axes[0, 0].set_ylim(0, 25)
axes[0, 0].annotate("Vertex", xy=(0, 0), xytext=(-3, 5), arrowprops=dict(facecolor='black', arrowstyle="->"))


axes[0, 1].hist(data, bins=30, color="red")
axes[0, 1].set_title("Histogram of Random Normal Values")
axes[0, 1].set_xlabel("Value")
axes[0, 1].set_ylabel("Frequency")


axes[1, 0].bar(categories, values, color="orange")
axes[1, 0].set_title("Bar Chart")
axes[1, 0].set_xlabel("Categories")
axes[1, 0].set_ylabel("Values")


scatter = axes[1, 1].scatter(x_scatter, y_scatter, c=colors, s=150, cmap="viridis", edgecolors="black")
axes[1, 1].set_title("Scatter Plot with Colormap & Size Encoding")
axes[1, 1].set_xlabel("X-axis")
axes[1, 1].set_ylabel("Y-axis")
axes[1, 1].set_xlim(0, 1)
axes[1, 1].set_ylim(0, 1)
fig.colorbar(scatter, ax=axes[1, 1], label="Color scale")

plt.tight_layout()
plt.show()
```
