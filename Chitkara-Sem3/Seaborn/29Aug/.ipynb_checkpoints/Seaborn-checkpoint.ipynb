```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def lineplot(n=3):
    x = np.linspace(0, 10, 100)  
    for i in range(1, n + 1):
        y = x * i   
        plt.plot(x, y, label=f"Line {i}")  
sns.set_theme()  

lineplot()


plt.legend()
plt.show()
sns.reset_orig()  
lineplot()
plt.title("Matplotlib defaults after sns.reset_orig()")
plt.show()
```
